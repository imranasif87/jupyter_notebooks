```python
import pandas as pd
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
import statistics
```

```python
data = pd.read_csv("evaluation_ims.csv") 
precision_ims = data['Precision'].tolist()
recall_ims = data['Recall'].tolist()
time_ims = data['time'].tolist()

precision_ims.sort()
recall_ims.sort()
time_ims.sort()

plt.step(recall_ims, precision_ims, color='g')
#plt.fill_between(recall, precision, alpha=0.2, color='g', step='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall curve')
plt.show()

print("Avg. Precision:" , statistics.mean(precision_ims))
print("Avg. Recall:" , statistics.mean(recall_ims))
```

```python
records = []
for i in range(1000):
    records.append(i)
```

```python
data = pd.read_csv("evaluation_lucene.csv") 
precision_lucene = data['Precision'].tolist()
recall_lucene = data['Recall'].tolist()
time_lucene = data['time'].tolist()

precision_lucene.sort()
recall_lucene.sort()
time_lucene.sort()

plt.step(recall_lucene, precision_lucene, color='g', alpha=0.2, where='post')
#plt.fill_between(recall, precision, alpha=0.2, color='g', step='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall curve')
plt.show()

print("Avg. Precision:" , statistics.mean(precision_lucene))
print("Avg. Recall:" , statistics.mean(recall_lucene))
```

```python
plt.figure(figsize=(18,8))
plt.step(recall_ims, precision_ims, color='r', label='Nanopub Retrieval Framework')
plt.step(recall_lucene, precision_lucene, color='g', label='Apache Lucene')
#plt.fill_between(recall, precision, alpha=0.2, color='g', step='post')
plt.xlabel('Recall', fontsize=18)
plt.ylabel('Precision', fontsize=18)
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('Precision-Recall curve', fontsize=18)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.step(time_ims, records, color='r', label='Nanopub Retrieval Framework')
plt.step(time_lucene, records, color='g', label='Apache Lucene')
#plt.fill_between(recall, precision, alpha=0.2, color='g', step='post')
plt.xlabel('Time (ms)', fontsize=18)
plt.ylabel('Search Terms', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('Time Execution on each search term', fontsize=18)
plt.legend()
plt.show()
```
